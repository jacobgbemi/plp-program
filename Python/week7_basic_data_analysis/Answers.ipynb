Create at least four different types of visualizations:

```python
import matplotlib.pyplot as plt
```

```python
salesmen["Revenue"].plot()
```

```python
salesmen["Revenue"].plot(kind="barh")
```

```python
salesmen["Revenue"].plot(kind="bar", legend=True)
```

```python
salesmen["Revenue"].plot(kind="pie")
```

```python
# Plotting Revenue vs. Salesman
df = salesmen
plt.plot(df['Salesman'], df['Revenue'])
plt.xlabel('Salesman')
plt.ylabel('Revenue')
plt.title('Salesman vs Revenue')
plt.show()
```

```python
# Plotting a bar chart for Revenue by Date
df = salesmen
df.groupby('Date')['Revenue'].mean().plot(kind='bar')
plt.xlabel('Date')
plt.ylabel('Average Revenue')
plt.title('Average Revenue by Date')
plt.show()
```

```python
# Plotting a bar chart for Revenue by Salesman
df = salesmen
df.groupby('Salesman')['Revenue'].mean().plot(kind='bar')
plt.xlabel('Salesman')
plt.ylabel('Average Revenue')
plt.title('Average Revenue by Salesman')
plt.show()
```

```python
# Plotting a bar chart for Revenue by Salesman
df = salesmen
df.groupby('Salesman')['Revenue'].mean().plot(kind='pie')
plt.xlabel('Salesman')
plt.ylabel('Average Revenue')
plt.title('Average Revenue by Salesman')
plt.show()
```

```python
# Plotting a histogram for Revenue
df = salesmen
df['Revenue'].plot(kind='hist', bins=10)
plt.xlabel('Revenue')
plt.title('Revenue Distribution')
plt.show()
```

```python
# Plotting a histogram for Revenue
df = salesmen
df.groupby("Salesman")['Revenue'].plot(kind='hist', bins=90, legend=True)
plt.xlabel('Revenue')
plt.title('Revenue Distribution')
plt.show()
```

```python
# Plotting a scatter plot of Revenue vs. Date
df = salesmen
df.plot(kind='scatter', x='Date', y='Revenue')
plt.title('Date vs Revenue')
plt.show()
```

```python
# Plotting a scatter plot of Revenue vs. Salesman
df = salesmen
df.plot(kind='scatter', x='Salesman', y='Revenue')
plt.title('Salesman vs Revenue')
plt.show()
```
